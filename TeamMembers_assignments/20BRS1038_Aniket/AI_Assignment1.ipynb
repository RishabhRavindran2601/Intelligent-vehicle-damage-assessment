```python
# Task 1
import numpy as np
import pandas as pd

np.random.seed(42) # Set the random seed

df = pd.DataFrame(np.random.rand(4, 4), columns=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'])
df
```

```python
# Task 2
new_column_names = {f'Feature {i+1}':f'Random value {i+1}' for i in range(4)}
df = df.rename(columns=new_column_names)
print(df)
```

```python
# Task 3
# Descriptive statistics
df.describe()
```

```python
# 
# Check for null values
null_values = df.isnull().sum()
print("Null values:\n" +str(null_values))

# Find the datatype of columns
data_types = df.dtypes
print("\nData Types:\n" +str(data_types))
```

```python
# Display columns using location method
loc_result = df.loc[:, ['Random value 2', 'Random value 3']]
print('Label-based location method (loc):\n',loc_result)

# Display columns using index location method
iloc_result = df.iloc[:, [1, 2]]
print("\nIndex-based location method (iloc):\n",iloc_result)
```
